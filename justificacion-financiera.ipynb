```python
# Justificación de Pérdidas por Cancelación y Potencial de Retención

import pandas as pd
import numpy as np

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn_Myrna.csv')

# Convertir a numérico 
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')

# Análisis de Pérdidas Actuales
churn_df = df[df['Churn'] == 'Yes']
clientes_perdidos = len(churn_df)
perdida_mensual = churn_df['MonthlyCharges'].sum()
perdida_anual = perdida_mensual * 12

print(f"--- ANÁLISIS DE PÉRDIDAS ACTUALES ---")
print(f"Clientes que cancelaron: {clientes_perdidos}")
print(f"Pérdida Mensual: ${perdida_mensual:,.2f}")
print(f"Pérdida Anual Proyectada: ${perdida_anual:,.2f}")
```

```python
# Justificación del Modelo y Retención
recall = 0.78  # 78% de Recall
tasa_recuperacion = 0.18     # 20% retención

# Clientes que el modelo logra identificar
clientes_detectados_val = perdida_anual * recall

# Dinero recuperado si ventas convence al 18% de esos detectados
dinero_recuperado = clientes_detectados_val * tasa_recuperacion

print(f"\n--- PROYECCIÓN DE RETENCIÓN ---")
print(f"Pérdida Anual Total: ${perdida_anual:,.2f}")
print(f"Pérdida Identificada por Modelo (78%): ${clientes_detectados_val:,.2f}")
print(f"Recuperación Estimada (18% de éxito en ventas): ${dinero_recuperado:,.2f}")
```
